```python
from tkinter import*

from tkinter import messagebox

from tkinter import ttk



root=Tk()

root.title('income tax calculator')

def func_income(sal,hra,var):
    
    if(sal==""or hra==""):
        messagebox.showerror("ERROR","PLEASE ENTER VALID CREDENTIALS")
    else:    
        messagebox.showinfo("INFO","YOUR DATA HAS BEEN SUBMITTED,CLICK NEXT TO PROCEED")
    
    
    global x,y,tax,z
    tax=0
    x=int(sal)
    y=int(hra)
    z=var
    
    if(y>x):
        messagebox.showerror("ERROR","PLEASE ENTER VALID CREDENTIALS")
   
    if(var==0):
        
        if x <= 250000:  #2 Lakh 50 thousand
            tax = 0

        elif x <= 500000: #5 Lakh
            tax = (x - 250000) * 0.05

        elif x <= 750000: #7 lakh 50 thousand
            tax = (x - 500000) * 0.10 + 12500 

        elif x <= 1000000: #10 Lakh
            tax = (x - 750000) * 0.15 + 37500 

        elif x <= 1250000: #12 lakh 50 thousand
            tax = (x- 1000000) * 0.20 + 75000 

        elif x<= 1500000:
            tax = (x - 1250000) * 0.25 + 125000 

        else:
            tax = (x - 1500000) * 0.30 + 187500

        
    
    elif(var==1):
        if x <= 300000:  #2 Lakh 50 thousand
            tax = 0

        elif x <= 500000: #5 Lakh
            tax = (x - 300000) * 0.05

        elif x <= 1000000: #7 lakh 50 thousand
            tax = (x - 500000) * 0.20 + 10000

        else: 
            tax = (x - 1000000) * 0.30 + 110000
        

    else:
        if x<= 500000:  #2 Lakh 50 thousand
            tax = 0

        elif x <= 500001: #5 Lakh
            tax = (x - 300000) * 0.05

        elif x <= 1000001: #7 lakh 50 thousand
            tax = (x - 500000) * 0.20 

        else: 
            tax = (x - 1000000) * 0.30 + 100000
            
    
        
    
   
    
    
    
    
def func(inv1,inv2,inv3,inv4):
    if(inv1==""or inv2=="" or inv3=="" or inv4==""):
        messagebox.showerror("ERROR","PLEASE ENTER VALID CREDENTIALS                                                                 (IF YOU DON'T HAVE ANY INVESTMENTS,ENTER 0 )")
    else:    
        messagebox.showinfo("INFO","YOUR DATA HAS BEEN SUBMITTED,CLICK CALCULATE TO PROCEED")
    global tax2
    tax2= 0
    a=int(inv1)
    b=int(inv2)
    c=int(inv3)
    d=int(inv4)
    
    
    
   #condition for investments
    if(a>150000):
        a=150000
    if(b>50000):
        b=50000
    if(d>25000):
        d=25000
    
    x2=(x-(a+b+d+y))
        
  #code for old regime

    if(z==0):
        if x2 <= 250000:  #2 Lakh 50 thousand
            tax2= 0

        elif x2 <= 500000: #5 Lakh
            tax2 = x*0.05

        elif x2 <= 1000000: #7 lakh 50 thousand
            tax2 = 12500+((x-500000)*0.2)

        elif x2 > 1000000: #10 Lakh
            tax2 = 112500+((1000000-x) * 0.3)
    elif(z==1):
        if x2 <= 300000:  #2 Lakh 50 thousand
            tax2 = 0

        elif x2 <= 500000: #5 Lakh
            tax2 = x*0.05

        elif x2 <= 1000000: #7 lakh 50 thousand
            tax2= 10000+((x-500000)*0.2)

        elif x2 > 1000000: #10 Lakh
            tax2 = 110000+((1000000-x)* 0.3)
    else:
        if x2 <= 500000:  #2 Lakh 50 thousand
            tax2 = 0

        elif x2 <= 1000000: #7 lakh 50 thousand
            tax2 = x*0.2

        elif x2 > 1000000: #10 Lakh
            tax2 = 100000+((1000000-x)* 0.3)
    tax2=tax2+c
    if (tax2<0):
        messagebox.showerror("ERROR","PLEASE ENTER VALID DATA")
    
             

def home():
    global f
    f=Frame(root,bg="orange")
    f.place(x="0",y="0",height="500",width="600")
    #label
    lbl=Label(f,text="Calculate income tax on your earnings in a few simple steps",bg="black",fg="white",font="times")
    lbl.place(x=20,y=7)
    #label for basic detail
    lbl=Label(f,text="BASIC DETAIL",bg="black",fg="white")
    lbl.place(x=10,y=54)
    #label for age
    age=Label(f,text="AGE",bg="black",fg="white")
    age.place(x=10,y=90)
    #label for salary
    salary=Label(f,text="SALARY",bg="black",fg="white")
    salary.place(x=10,y=220)
    #label for HRA
    hra=Label(f,text="HOUSE RENT ALLOWANCE",bg="black",fg="white")
    hra.place(x=10,y=280)
    #label for any other invesment
    inv=Label(f,text="GENDER",bg="black",fg="white")
    inv.place(x=10,y=160)
    #age
    var=IntVar()
    
    radio=Radiobutton(root,text="Below 60 years",width="15",value=0,variable=var)
    radio.place(x=10,y=120)
    radio=Radiobutton(root,text="61-80 years",width="15",value=1,variable=var)
    radio.place(x=200,y=120)
    radio=Radiobutton(root,text="Above 81",width="15",value=2,variable=var)
    radio.place(x=400,y=120)
    #entry box for salary
    sal=Entry(root)
    sal.place(x=10,y=250)
    #entry box for hra
    hra=Entry(root)
    hra.place(x=10,y=310)
    #combobox for investment
    com=ttk.Combobox(root,width=20)
    com['state']='readonly'
    com['values']=('MALE','FEMALE' )
    com.current(0)
    com.place(x=10,y=190)
    for_submit=Button(f,text="SUBMIT",command=lambda:func_income(sal.get(),hra.get(),var.get()))
    for_submit.place(x="100", y="450")
    home_to_next=Button(f,text="NEXT",command=next)
    home_to_next.place(x="200", y="450")

def next():
    f.place_forget
    global f2
    f2=Frame(root,bg="#49A")
    f2.place(x="0",y="0",height="500",width="600")
    mylable=Label(f2,text="INVESTMENT DETAILS")
    mylable.place(x=10,y=30)
    inv=Label(f2,text="Investments u/s 80C of Income Tax Act 1961"
         ,bg="black",fg="white")
    inv.place(x=10,y=60)
 
    inv=Label(f2,text="Investments u/s 80CCD(1B) of Income Tax Act 1961"
         ,bg="black",fg="white")
    inv.place(x=10,y=100)
    inv=Label(f2,text="Professional Tax"
         ,bg="black",fg="white")
    inv.place(x=10,y=160)
    inv=Label(f2,text="Medical insurance Premium deduction u/s 80D of the Income Tax Act 1961"
         ,bg="black",fg="white")
    inv.place(x=10,y=225)
    #entry box
   
    inv1=Entry(f2)
    inv1.place(x=10,y=83)
    inv2=Entry(f2)
    inv2.place(x=10,y=135)
    inv3=Entry(f2)
    inv3.place(x=10,y=190)
    inv4=Entry(f2)
    inv4.place(x=10,y=250)
    #submission
 
    
    submit=Button(f2,text="SUBMIT",command=lambda:func(inv1.get(),inv2.get(),inv3.get(),inv4.get()))
    submit.place(x="100",y="450")
    
    next_tonext=Button(f2,text="CALCULATE",command=last)
    next_tonext.place(x="200", y="450")
    
def last():
    f2.place_forget()
    global f3
    f3=Frame(root,bg="Pink3")
    f3.place(x="0",y="0",height="500",width="600")
    mylable=Label(f3,text="gross income")
    f3.place(x="0",y="0",height="500",width="600")
    mylable=Label(f3,text="TAX ACC. TO NEW REGIME")
    mylable.place(x=10,y=30)
    mylable=Label(f3,text=tax)
    mylable.place(x=170,y=30)
    mylable=Label(f3,text="TAX ACC. TO OLD REGIME")
    mylable.place(x=10,y=70)
    mylable=Label(f3,text=tax2)
    mylable.place(x=170,y=70)
    
    
    
button = Button(root, text="Start", padx=50, pady=10,  command=home ,bg="#e09e50")
button.place(x="245", y="225")
lbl=Label(root,text="Calculate income tax on your earnings in a few simple steps",bg="black",fg="white",font="times")
lbl.place(x=20,y=155)
root.configure(background="pink")
root.geometry("600x500")
root.maxsize(height="500",width="600")

root.mainloop()
```
